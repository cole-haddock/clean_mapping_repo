```python
import pandas as pd
```

```python
df = pd.read_csv('../data/sweep_df.csv')
df.head(2)
```

```python
df['postings_per_sweep'] = df.groupby('sweep_event_id')['sweep_event_id'].transform('count')
df = df.sort_values('postings_per_sweep', ascending=False)
df.head(2)
```

```python
sweeps_table = (
    df.groupby('sweep_event_id')
    .agg(
        postings_per_sweep=('postings_per_sweep', 'first')
    )
    .sort_values('postings_per_sweep', ascending=False)
    .reset_index()
)

print(sweeps_table.head(25))
```
